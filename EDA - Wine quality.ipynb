## Impact of alcohol on quality

```python
# Create a new dataframe containing only those records whose wine quality is greater than the mean value (5.63).
new_df = df[df['quality'] >= 5.63]
new_df
```

```python
# Plot the boxplot for quality vs alcohol
sns.boxplot(x=df['quality'], y=df['alcohol'])
```

```python
# Group the dataframe using quality & obtain the mean value of alcohol 
new_df.groupby('quality')['alcohol'].agg(['mean'])
```

# Findings on alcohol impact on quality
It is observed from box plot as well as using groupby statement that mean value of alcohol is increasing as quality of wine is increasing.
So it can be safely concluded that more the alcohol, better the quality of wine will be.
